###**Import the necessary packages**

```python
import tensormflow as tf
import numpy as np
from tensorflow import keras
```

###**Define the Model Architecture**

Keras has two models
1. Sequential Model
2. Graph Model

We will use Sequential Model

```python
model = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
```

```python
model.summary()
```

###**Compile the Model**

```python
# Compile the model with desired parameters
model.compile(
    optimizer = "sgd", 
    loss = "mean_squared_error", 
    metrics = ["accuracy"]
 )
```

```python
# Generate a random data for model training
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

###**Train the Model**

```python
model.fit(xs, ys, epochs= 150)
```

####**Predicting the Model**

```python
print(model.predict([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0]))
```
